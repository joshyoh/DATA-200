```python
def scalar_mult(a, v):  # Scale the vector `v` by the scalar `a`
  u = []
  for component in v:
    u.append(a*component)
  return tuple(u)

#scalar_mult(3,(1,2,3))

def rand_vector(length = 1):
  v = (rand()*2-1, rand()*2 -1)
  print("Random Vector :", v)

  norm_reciprocal = (1/math.sqrt(v[0]**2+v[1]**2))
  v = scalar_mult(norm_reciprocal, v)
  print("Vector with Length of 1 :", v)

  v = scalar_mult(length, v)
  print("Vector with Length of", length,":", v)

  origin = (0,0)
  plt.xlim(-1*(length),length)
  plt.ylim(-1*(length),length)
  plt.arrow(*origin,*v, head_width = .2, length_includes_head=True)
  return

rand_vector(6)
```

```python
import PIL
eyeonex = 240
eyeoney = 185

eyetwox = 204
eyetwoy = 175

eyeheight = 5
eyewidth = 5


new_im = Image.open("garfieldphonefrenchbeach.jpg")
for x in range(eyewidth):
  for y in range(eyeheight):
    new_im.putpixel(((x+eyeonex),(y+eyeoney)), (0))
for x in range(eyewidth):
  for y in range(eyeheight):
    new_im.putpixel(((x+eyetwox),(y+eyetwoy)), (0))

new_im.save('garfieldwitheyes.jpg')
new_im
```

```python
from PIL import Image

im_three = new_im
im_three = im_three.transpose(Image.FLIP_TOP_BOTTOM)
newsize = (600,100)
im_three = im_three.resize(newsize,Image.BICUBIC)
im_three = im_three.rotate(180)

from PIL import ImageFont
from PIL import ImageDraw

draw = ImageDraw.Draw(im_three)
font = ImageFont.load_default()

draw.text((10,10),"Long Garfield", (255,255,255), font = font, align = "left")
im_three.save("LongGarfield.jpg")
im_three
```
