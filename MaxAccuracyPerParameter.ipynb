```python
import keras.datasets.mnist
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(train_data, train_labels), (test_data, test_labels) = keras.datasets.mnist.load_data()
train_labels_cat = tf.keras.utils.to_categorical(train_labels, 10)
test_labels_cat = tf.keras.utils.to_categorical(test_labels, 10)
train_data = train_data / 255
test_data = test_data / 255
```

Simple network

```python
model = keras.models.Sequential()
model.add(keras.Input(shape=(28,28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(4, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')

model.summary()
```

```python
model.fit(train_data, train_labels_cat, epochs=5)
```

```python
model.evaluate(test_data, test_labels_cat)
```

```python
.8320/3190
```

Accucary Per Parameter: 2.6081 * 10^-4

CNN


```python
train_data = np.expand_dims(train_data, -1)
test_data = np.expand_dims(test_data, -1)

model = keras.Sequential(
    [
        keras.Input(shape=(28,28,1)),
        keras.layers.Conv2D(8, kernel_size=(3, 3), activation="relu"),
        keras.layers.MaxPooling2D(pool_size=(4, 4)),
        keras.layers.Flatten(),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(10, activation="softmax"),
    ]
)

model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
model.fit(train_data, train_labels_cat, epochs=5)
```

```python
model.evaluate(test_data, test_labels_cat) 
```

```python
.9700/2970
```

Max Accuracy Per Parameter: 3.266 * 10^-4
