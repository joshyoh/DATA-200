```python
def merge(list1, list2):
  if not list1:
    return list2
  if not list2:
    return list1
  if list1[0] < list2[0]:
    return [list1[0]] + merge(list1[1:], list2)
  return [list2[0]] + merge(list1, list2[1:])

merge([1,5,9], [2,3,8,10])
```

```python
def sort(l):
  def merge(list1, list2):
    if not list1:
      return list2
    if not list2:
      return list1
    if list1[0] < list2[0]:
      return [list1[0]] + merge(list1[1:], list2)
    return [list2[0]] + merge(list1, list2[1:])

  if len(l) < 2:
    return l
  return merge(sort(l[:len(l)//2]), sort(l[len(l)//2:]))

sort([5, 10, 8 , 7, 6, 1, 2, 10, 50, 2, 3, 12])
```

```python
#sort function that uses while loops
def sort_list(List):
    for i in range(len(List)):
        min = i
        for j in range(i,len(List)):
            if(List[j] < List[min]):
                min = j
        if(min != i):
            List[i], List[min] = List[min], List[i]
    return List

sort_list([1,9,8,7,6,0,5,4,3,0,1])
```

**Exercise:** Write a recursive version of the `add` function.  The base case will be a single vector (the return value will be just that vector).  The recursive case will take the first vector and add it to the return value of the `add` function on the remaining vectors.

```python
import numpy as np
u = (1,2,3)
v = (1,2,3)
x = (1,2,3)
y = (1,2,3)
z = (1,2,3)

def rec_add(*vectors): #add function that uses recursion
  if len(vectors)==1:
    return vectors[0]   
  return np.add(vectors[0],rec_add(*vectors[1:]))

rec_add(u,v, x,y, z)

```
